```python
def print_chessboard(board, n):
    print("+" + "---+" * n)
    
    for row in board:
        line = "|"
        for cell in row:
            line += f" {cell} |"
        print(line)
        print("+" + "---+" * n)
    
    print()

def solve_n_queens(n):
    if n < 1:
        print("N must be at least 1.")
        return
    
    col_used   = [False] * n
    diag1_used  = [False] * (2 * n - 1) 
    diag2_used = [False] * (2 * n - 1)
    
    board = [['x' for _ in range(n)] for _ in range(n)]

    def backtrack(row):
        if row == n:
            print(f"\nSolution for N_Queens = {n}:")
            print_chessboard(board, n)
            return True 

        for c in range(n):
            d1 = row + c
            d2 = row - c + n - 1
            
            if not col_used[c] and not diag1_used[d1] and not diag2_used[d2] :

                board[row][c]= 'Q'
                col_used[c] = diag1_used[d1] =diag2_used[d2] = True
                
                if backtrack(row + 1):
                    return True

                board[row][c] = 'x'
                col_used[c] =diag1_used[d1] = diag2_used[d2] = False
        
        return False

    if not backtrack(0):
        print(f"No solution exists for N = {n}")

try:
    n = int(input("Enter N (number of queens and board size): "))
    solve_n_queens(n)
except ValueError:
    print("Please enter a valid integer.")
```
