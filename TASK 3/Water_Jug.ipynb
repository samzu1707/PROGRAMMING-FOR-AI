```python
class WaterJugDFS:
    def __init__(self, jug1_capacity=4, jug2_capacity=3, goal=2):
        self.jug1_capacity = jug1_capacity
        self.jug2_capacity = jug2_capacity
        self.goal = goal
        self.visited = set()

    def is_goal(self, state):
        x, y = state
        return x == self.goal or y == self.goal

    def get_next_states(self, state):
        x, y = state
        next_states = []

        next_states.append(((self.jug1_capacity, y), f"Fill Jug X -> ({self.jug1_capacity}, {y})"))
        next_states.append(((x, self.jug2_capacity), f"Fill Jug Y -> ({x}, {self.jug2_capacity})"))

        next_states.append(((0, y), f"Empty Jug X -> (0, {y})"))
        next_states.append(((x, 0), f"Empty Jug Y -> ({x}, 0)"))

        transfer = min(x, self.jug2_capacity - y)
        next_states.append(((x - transfer, y + transfer),
                            f"Pour from X until X is empty / Y not full -> ({x - transfer}, {y + transfer})"))

        if x + y > self.jug2_capacity:
            next_states.append(((x - (self.jug2_capacity - y), self.jug2_capacity),
                                f"Pour from X until Y is full -> ({x - (self.jug2_capacity - y)}, {self.jug2_capacity})"))
            
        transfer = min(y, self.jug1_capacity - x)
        next_states.append(((x + transfer, y - transfer),
                            f"Pour from Y until Y is empty / X not full -> ({x + transfer}, {y - transfer})"))

        if x + y > self.jug1_capacity:
            next_states.append(((self.jug1_capacity, y - (self.jug1_capacity - x)),
                                f"Pour from Y until X is full -> ({self.jug1_capacity}, {y - (self.jug1_capacity - x)})"))

        return next_states

    def dfs(self, current_state, path):
        if current_state in self.visited:
            return None
        self.visited.add(current_state)

        if self.is_goal(current_state):
            return path

        for next_state, rule in self.get_next_states(current_state):
            if next_state not in self.visited:
                result = self.dfs(next_state, path + [rule])
                if result:
                    return result
        return None

solver = WaterJugDFS()
solution = solver.dfs((0, 0), ["Start from (0, 0)"])

if solution:
    print("\nSolution Found:\n")
    for step in solution:
        print(step)
else:
    print("No solution found.")
```
